```python
#Question 4(I)
import matplotlib.pyplot as plt
import math as m

def exponent(x):
    Term_no =[]
    Term_value=[]
    True_relative_error = []
    Approximate_relative_error = []
    
    y=f'{m.exp(-5):.7}'
    print('True value of 𝑒^(−5)=',y)
    
    
    j=0
    t=0.0
    h=0.0
    p=0.0
    q2=0.0
    while j<=19:
        g=(m.pow(x,j))/(m.factorial(j))
        t=t+g
        
        q1=h
        if(x<0):
             h=f'{t:.7}'
              
        else:
            h=f'{(1/t):.7}'
            
        
        p=f'{(((float(h)-float(y))/float(y)))*100:.7}'
        
        if(q1!=0):
            q2=f'{(abs((float(h)-float(q1))/float(h))*100):.7}'
        
        
    
        Term_no.append(j+1)
        Term_value.append(h)
        True_relative_error.append(p)
        Approximate_relative_error.append(q2)
                
        j=j+1 
    
    
    titles = ['Term no', 'Term value', 'True relative error (𝜀𝑡 %)', 'Approximate relative error (𝜀𝑎 %)']
    data = [titles] + list(zip(Term_no, Term_value , True_relative_error,Approximate_relative_error))

    for i, d in enumerate(data):
        line = '|'.join(str(a).ljust(27) for a in d)
        print(line)
        if i == 0:
            print('-' * len(line))
    
    return h

print("I)\n")
print("Case a")
z=exponent(-5)

print("\nCase b")
w=1/float(exponent(5))
```

```python
#Question 4(II & III)
import math
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

e_true = float(f'{math.exp(-5):.7}')
def func_exp(x,n):
    t= 0
    e_approx=0
    for i in range(n):
        num = x**i
        denom = math.factorial(i)
        t += ( (num)/(denom) )
        
        
        if(x<0):
            e_approx = float(f'{t:.7}')
        else:
            e_approx = float(f'{(1/t):.7}')
        
    return e_approx

def func_trerror(x,x_value):
    e_approx = func_exp(x,x_value)
    e_trerror =float(f'{(((e_approx - e_true)/e_true))*100:.7}')
    #print(f'{x_value} terms: Taylor Series approx= {e_approx}, exp(-5)= {e_true}, trerror = {e_trerror}')
        
    return e_trerror

def func_aprerror(x,x_value):
    e_approx_old= func_exp(x,x_value-1)
    e_approx_new= func_exp(x,x_value)
    
    if(e_approx_old==0):
        e_aprerror=0.0
    else:
        e_aprerror =float(f'{(abs((e_approx_new - e_approx_old)/e_approx_new))*100:.7}')
    #print(f'{x_value} terms: Taylor Series approx= {e_approx_new},\t aprerror={e_aprerror}')
        
    return e_aprerror


x_values = np.arange(1,21,1)

#Question 4(II)
trerrora = [func_trerror(-5,x_value) for x_value in x_values]
trerrorb = [func_trerror(5,x_value) for x_value in x_values]

plt.figure(figsize=(15,6))
plt.plot(x_values,trerrora,'b',label='case a')
plt.plot(x_values,trerrorb,'r',label='case b')

plt.title("Variation of True Relative Error %",fontsize=20)
plt.xlabel("Term no",fontsize=18)
plt.ylabel("True relative error (%)",fontsize=18)
plt.legend(loc=1)
plt.grid()

#Question 4(III)
aprerrora = [func_aprerror(-5,x_value) for x_value in x_values]
aprerrorb = [func_aprerror(5,x_value) for x_value in x_values]

plt.figure(figsize=(15,6))
plt.plot(x_values,aprerrora,'b',label='case a')
plt.plot(x_values,aprerrorb,'r',label='case b')

plt.title("Variation of Aprroximate Relative Error %",fontsize=20)
plt.xlabel("Term no",fontsize=18)
plt.ylabel("Approximate relative error (%)",fontsize=18)
plt.legend(loc=1)

plt.grid()
plt.show()


#Question 4(IV)
print('IV )\n𝜀𝑠 = 0.000005')
y_values = np.arange(2,30,1)
for y_value in y_values:
    eaa = func_aprerror(-5,y_value)
    #print(y_value,eaa)
    if(eaa<0.000005):
        print("No. of terms required for 𝜀𝑎 < 𝜀𝑠 in case a = ",y_value)
        break   
#print('\n')
for y_value in y_values:
    eab =func_aprerror(5,y_value) 
    #print(y_value,eab)
    if(eab<0.000005):
        print("\nNo. of terms required for 𝜀𝑎 < 𝜀𝑠 in case b = ",y_value)
        break
```
